```python
import pandas as pd
```

```python
input_file = 'input.csv'
output_file = 'output.csv'
```

```python
df = pd.read_csv(input_file)
```

```python
df
```

```python
transformed_data = []

for index, row in df.iterrows():
    emp_code = row['Employee Code']
    manager_emp_code = row['Manager Employee Code']
    
    for i in range(1, 3):
        compensation_col = f'Compensation {i}'
        compensation_date_col = f'Compensation {i} date'
        review_col = f'Review {i}'
        review_date_col = f'Review {i} date'
        
        # Check if compensation data is not NaN
        if not pd.isna(row[compensation_col]):
            effective_date = row[compensation_date_col]
            end_date = '2100-01-01'  # Default end date
            
            if index < len(df) - 1:
                next_row = df.iloc[index + 1]
                next_effective_date = next_row[review_date_col]
                end_date = pd.to_datetime(next_effective_date) - pd.DateOffset(days=1)
                end_date = end_date.strftime('%Y-%m-%d')
            
            transformed_data.append({
                'Employee Code': emp_code,
                'Manager Employee Code': manager_emp_code,
                'Last Compensation': row[compensation_col],
                'Compensation': row[compensation_col],
                'Last Pay Raise Date': row[compensation_date_col],
                'Variable Pay': 0.0,  # Assuming variable pay is not available
                'Tenure in Org': 0.0,  # Assuming tenure is not available
                'Performance Rating': row[review_col],
                'Engagement Score': row[review_col],
                'Effective Date': effective_date,
                'End Date': end_date
            })
```

```python
transformed_df = pd.DataFrame(transformed_data)
```
